```python
import math
import random
```

```python
# Lista de nÃºmeros aleatorios
random_numbers = [random.randint(0, 20) for _ in range(100)]
```

```python
# Calcular la probabilidad de cada numero
probabilities = [random_numbers.count(i) / len(random_numbers) for i in range(1, 21)]
```

```python
# Calcular entropia
entropy = -sum(p * math.log2(p) for p in probabilities if p > 0)
entropy
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(list(range(1, 21)), probabilities)
plt.xlabel('number')
plt.ylabel('probability')
plt.show
```
